```python
import os, getpass

from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.messages import get_buffer_string
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.runnables.history import RunnableWithMessageHistory

from langgraph.store.base import BaseStore
from langchain_core.runnables.config import RunnableConfig
```

```python
# 2. Initialize the updated history object
history = InMemoryChatMessageHistory()

# Add user and AI messages
history.add_user_message("I'm heading to New York next week.")
history.add_ai_message("Great! It's a fantastic city.")

# Access the list of messages
print(history.messages)
```

```python
# 1. Initialize the modern memory object
memory = InMemoryChatMessageHistory()

# 2. Save a conversation turn (Replaces save_context)
memory.add_user_message("What's the weather like?")
memory.add_ai_message("It's sunny today.")

# 3. Load the memory as a string (Replaces load_memory_variables)
# get_buffer_string automatically formats the messages into the classic "Human/AI" text block
buffer_string = get_buffer_string(memory.messages)

print({"history": buffer_string})
```

```python
# 1. Define LLM and Prompt
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# We use ChatPromptTemplate with a MessagesPlaceholder to dynamically inject history
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful travel agent."),
    MessagesPlaceholder(variable_name="history"), # The memory goes here
    ("human", "{question}")
])

# 2. Build the core chain (No memory attached yet)
chain = prompt | llm

# 3. Configure Memory Storage
# We create a dictionary to hold the chat history for different sessions/users
store = {}

def get_session_history(session_id: str) -> InMemoryChatMessageHistory:
    """Fetches or creates a new memory object for a given session."""
    if session_id not in store:
        store[session_id] = InMemoryChatMessageHistory()
    return store[session_id]

# 4. Wrap the chain with the History Manager
conversational_chain = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="question", # Matches the human prompt variable
    history_messages_key="history", # Matches the placeholder variable
)
```

```python

# 5. Run the Conversation
# You MUST pass a config dictionary with a session_id so it knows whose memory to read
config = {"configurable": {"session_id": "sam_session_001"}}

print("--- Turn 1 ---")
response = conversational_chain.invoke({"question": "I want to book a flight."}, config=config)
print(response.content)

print("\n--- Turn 2 ---")
response = conversational_chain.invoke({"question": "My name is Sam, by the way."}, config=config)
print(response.content)

print("\n--- Turn 3 ---")
response = conversational_chain.invoke({"question": "What was my name again?"}, config=config)
print(response.content)
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a friendly assistant."),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{question}")
])

# 3. Build the core chain
chain = prompt | llm

# 4. Configure Memory Storage
store = {}

def get_session_history(session_id: str) -> InMemoryChatMessageHistory:
    if session_id not in store:
        store[session_id] = InMemoryChatMessageHistory()
    return store[session_id]

# 5. Wrap the chain with the History Manager
conversational_chain = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="question",
    history_messages_key="chat_history",
)

# 6. Run the Conversation
config = {"configurable": {"session_id": "jane_session_001"}}

print("--- Turn 1 ---")
response = conversational_chain.invoke({"question": "Hi, I'm Jane."}, config=config)
print(response.content)

print("\n--- Turn 2 ---")
response = conversational_chain.invoke({"question": "Do you remember my name?"}, config=config)
print(response.content)
```

```python
def update_instructions(state: dict, config: RunnableConfig, store: BaseStore):
    # 1. Fetch user_id from config so instructions are personalized
    user_id = config.get("configurable", {}).get("user_id", "default_user")
    
    # 2. Define a consistent namespace
    namespace = ("agent_instructions", user_id)
    
    # 3. SAFELY get current instructions (No [0]!)
    existing_item = store.get(namespace, key="agent_a")
    
    # 4. Fallback if the store is empty (First run)
    if existing_item:
        current_instructions = existing_item.value.get("instructions", "You are a helpful assistant.")
    else:
        current_instructions = "You are a helpful assistant."

    # --- LLM INVOCATION LOGIC ---
    # prompt = prompt_template.format(
    #     instructions=current_instructions,
    #     conversation=state["messages"]
    # )
    # output = llm.invoke(prompt)
    # new_instructions = output['new_instructions']
    
    # Mocking the new instructions for this example
    new_instructions = "You are a helpful assistant who loves emojis."
    # ----------------------------

    # 5. Save the updated instructions back to the exact same namespace
    store.put(namespace, "agent_a", {"instructions": new_instructions})
    print(f"✅ Updated instructions for {user_id}: {new_instructions}")
    
    return state # LangGraph nodes should return the state update


# Node that uses the instructions to generate a response
def call_model(state: dict, config: RunnableConfig, store: BaseStore):
    # 1. Match the exact same namespace
    user_id = config.get("configurable", {}).get("user_id", "default_user")
    namespace = ("agent_instructions", user_id)
    
    # 2. SAFELY retrieve the item (No [0]!)
    item = store.get(namespace, key="agent_a")

    # 3. Fallback for the very first execution
    if item and "instructions" in item.value:
        instructions = item.value["instructions"]
    else:
        instructions = "You are a helpful assistant."

    print(f"🧠 Agent is using these instructions: '{instructions}'")

    # Use the retrieved instructions to format the prompt
    # prompt = prompt_template.format(instructions=instructions)
    # ... application logic continues ...
    
    return state
```

```python
from langgraph.store.memory import InMemoryStore

# A placeholder for a real embedding function
def embed(texts: list[str]) -> list[list[float]]:
    # In a real application, use a proper embedding model
    return [[1.0, 2.0] for _ in texts]

# Initialize an in-memory store. For production, use a database-backed store.
store = InMemoryStore(index={"embed": embed, "dims": 2})

# Define a namespace for a specific user and application context
user_id = "my-user"
application_context = "chitchat"
namespace = (user_id, application_context)

# 1. Put a memory into the store
store.put(
    namespace,
    "a-memory",  # The key for this memory
    {
        "rules": [
            "User likes short, direct language",
            "User only speaks English & python",
        ],
        "my-key": "my-value",
    },
)

# 2. Get the memory by its namespace and key
item = store.get(namespace, "a-memory")
print("Retrieved Item:", item)

# 3. Search for memories within the namespace, filtering by content
# and sorting by vector similarity to the query.
items = store.search(
    namespace,
    filter={"my-key": "my-value"},
    query="language preferences"
)
print("Search Results:", items)
```
